#### Pseudocode for  project 1


##### INPUT number
##### SUBTRACT 17 from number
##### IF number > 17 return double the difference
##### IF number < 17 return the absolute difference


```python
x = int(input("type in a number"))
y = x-17
if x>17:
    print (y*2)
elif x<17:
    print (abs(y))
```

#### Pseudocode for project 3

```python
a = int(input("type in a number"))
b = int(input("type in a number"))
if a == b:
    print ("True")
elif a+b == 5:
    print ("True")
elif a-b == 5:
    print ("True")
else:
    print ("False")
```
